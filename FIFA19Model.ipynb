```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('data.csv')
df.head()
```

```
df.drop(['Unnamed: 0', 'ID', 'Photo','Flag','Club Logo','Preferred Foot', 'International Reputation', 'Weak Foot',
       'Skill Moves', 'Work Rate', 'Body Type', 'Real Face', 'Position',
       'Jersey Number', 'Joined', 'Loaned From', 'Contract Valid Until','LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW',
       'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM',
       'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB'],axis=1,inplace=True)
```

```
df.columns
```

```
def club(x):
    return df[df['Club']==x][['Name','Age','Overall','Club','Value']]
```

```
juventus = club('Juventus')
real_madrid = club('Real Madrid')
fc_barcelona = club('FC Barcelona')
psg = club('Paris Saint-Germain')
atletico_madrid = club('Atlético Madrid')
man_city = club('Manchester City')
bayern_munchen = club('FC Bayern München')
man_united = club('Manchester United')
```

```
teams = ['juventus','real_madrid','fc_barcelona','psg','man_city','bayern_munchen','man_united']
print(teams)
```

```
def overall_sum(x):
    sum = 0
    for i in range (0,15):
        sum += x.iloc[i]['Overall']
    print(sum/15)

def overall_age(x):
    age = 0
    for i in range (0,15):
        age+=x.iloc[i]['Age']
    print(age/15)
```

```
import plotly.express as px
import plotly.graph_objects as go
import plotly.offline as py

def overall_graph(x):
    tempdf = x.sort_values(by='Overall')
    fig = px.bar(tempdf, x='Name', y='Overall', color='Overall')
    fig['layout']['yaxis1'].update(title='', range=[60, 100], dtick=5, autorange=False)
    fig.update_layout(title='OVR of Players',
                 xaxis_title="Player's Name",
                 yaxis_title="OVR")
    py.iplot(fig)
```
