```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
df.columns
```

```python
x = df.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans
```

```python
elb = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(x)
    elb.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),elb)
plt.title('The Elbow Method')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5,init='k-means++',max_iter=300,n_init=10,random_state=0)
```

```python
predict= kmeans.fit_predict(x)
```

```python
plt.scatter(x[predict==0,0],x[predict==0,1],s=100,c='red',label='1')
plt.scatter(x[predict==1,0],x[predict==1,1],s=100,c='blue',label='2')
plt.scatter(x[predict==2,0],x[predict==2,1],s=100,c='yellow',label='3')
plt.scatter(x[predict==3,0],x[predict==3,1],s=100,c='green',label='4')
plt.scatter(x[predict==4,0],x[predict==4,1],s=100,c='cyan',label='5')
plt.legend()
plt.show()
```
